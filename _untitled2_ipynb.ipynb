Типы графиков и диаграмм :

●	Диаграмма рассеяния — отображение взаимосвязи между двумя переменными. Каждый объект данных представлен точкой на графике, где координаты точки соответствуют значениям этих переменных.

●	Столбчатая диаграмма — визуализация категориальных данных. Она отображает значения в виде прямоугольников, высота которых соответствует количеству или величине для каждой категории. Столбчатые диаграммы подходят для визуализации динамики, рейтингов, соотношений между несколькими величинами, например выручки в двух магазинах.

●	Круговая диаграмма — визуализация пропорций категориальных данных. Она показывает доли каждой категории от общего количества. Круговая диаграмма подходит для визуализации, например, доли на рынке, продаж, опрошенных.

●	Линейный график — визуальное представление данных, где точки данных соединены линиями. Он обычно используется для отображения изменений значений переменной по отношению к другой переменной, например времени. Линейные графики подходят, например, для анализа трендов или изменения котировок ценных бумаг.

●	3D-графики — визуализация трёхмерных объектов и сцен с помощью компьютерной графики. Такие графики подходят, например, для визуализации реалистичных персонажей и окружения, объектов.

#**4. Обзор библиотек визуализации данных**

Matplotlib — это библиотека с открытым исходным кодом для визуализации данных в Python. С помощью неё можно создавать точечные и круговые диаграммы, линейные графики, гистограммы, диаграммы ошибок, 3D-графики. Готовые результаты визуализации можно экспортировать в форматах PDF, SVG, JPG, PNG, BMP и GIF.

Для работы с Matplotlib не требуются сложные протоколы. Библиотека помогает визуализировать большие объёмы данных с помощью коротких команд. Поэтому Matplotlib используют как замену MATLAB, более сложного инструмента для визуализации.

```python
import matplotlib.pyplot as plt

# Пример данных
categories = ['Январь', 'Февраль', 'Март','Апрель', 'Май', 'Июнь',
    'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']# Категория данных
values = [5, 22, 19, 51, 48, 53, 57, 60, 56, 49, 44, 39] #Значения показателей

plt.bar(categories, values)#Выбор типа диаграммы
plt.xlabel('Месяц') #Подпись для оси х
plt.ylabel('Показатель') #Подпись для оси y
plt.title('Столбчатая диаграмма   с Matplotlib')# Наименование диаграммы
plt.show()
```

Еще один пример круговой диаграммы при помощи редактора matplotlib.

```python
import matplotlib.pyplot as plt

vals = [5, 22, 19, 51, 48, 53, 57, 60, 56, 49, 44, 39] #Значения показателей
labels = ['Январь', 'Февраль', 'Март','Апрель', 'Май', 'Июнь',
    'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']# Категория данных

plt.pie(vals, labels=labels, autopct='%1.1f%%')
plt.title("Круговая диаграмма при помощи matplotlib")
plt.show()
```

Seaborn — это библиотека для создания статистической графики на Python. Она основана на matplotlib и тесно интегрируется со структурами данных pandas.

Seaborn помогает изучать и понимать ваши данные. Его функции построения графиков работают с фреймами данных и массивами, содержащими целые наборы данных, и выполняют необходимое семантическое сопоставление и статистическую агрегацию для создания информативных графиков. Его декларативный API, ориентированный на наборы данных, позволяет сосредоточиться на том, что означают различные элементы ваших графиков, а не на деталях их построения.

Вот пример того, что может сделать сиборн:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# Применить тему по умолчанию
sns.set_theme()

# Загрузить пример набора данных
tips = sns.load_dataset("tips")

# Создание визуализации
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
# Демонстрация
plt.show()
```

Еще один пример линейного графика при помощи библиотеки seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

Plotly — это библиотека визуализации данных с открытым исходным кодом, предоставляющая широкий спектр возможностей интерактивного построения графиков на языке Python. Plotly позволяет создавать интерактивные визуализации
Благодаря своей гибкости и простоте использования библиотека широко используется в различных областях: анализ данных, научные исследования, финансы, веб-разработка и других.

```python
import plotly

import plotly.graph_objs as go

import plotly.express as px

from plotly.subplots import make_subplots



import numpy as np

import pandas as pd

categories = ['Январь', 'Февраль', 'Март','Апрель', 'Май', 'Июнь',
    'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']# Категория данных
values = [5, 22, 19, 51, 48, 53, 57, 60, 56, 49, 44, 39] #Значения показателей


fig = go.Figure()
fig.add_trace(go.Scatter(x=categories, y=values))
fig.show()
```

```python
import plotly.express as px
df = px.data.gapminder().query("country == 'Canada'")
fig = px.bar(df, x='year', y='pop',
             hover_data=['lifeExp', 'gdpPercap'], color='lifeExp',
             labels={'pop':'population of Canada'}, height=400)
fig.show()
```

```python
import plotly


categories = ['Январь', 'Февраль', 'Март','Апрель', 'Май', 'Июнь',
    'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']# Категория данных
values = [5, 22, 19, 51, 48, 53, 57, 60, 56, 49, 44, 39] #Значения показателей

fig = go.Figure()
fig.add_trace(go.Pie(values=values, labels=categories))
fig.show()
```
